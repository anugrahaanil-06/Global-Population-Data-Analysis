```python
df.hist(figsize=(10,8), bins=10, edgecolor='black')
plt.suptitle("Distribution of Numerical Features", y=1.02)
plt.show()
```

```python
top_pop = df.nlargest(10, 'Population(in millions)')

plt.figure(figsize=(10,5))
sns.barplot(x='Country', y='Population(in millions)', data=top_pop, palette='viridis')
plt.title("Top 10 Countries by Total Population", fontsize=13, fontweight='bold')
plt.ylabel("Population (in millions)")
plt.xlabel("Country")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
top_pop.plot(
    x='Country',
    y=['Population Aged 0 to 14 (%)', 'Population Aged 60 and Over (%)'],
    kind='bar',
    color=['#1f77b4','#ff7f0e'],
    edgecolor='black'
)
plt.title("Age Group Comparison (0â€“14 vs 60+) - Top 10 Countries", fontsize=13, fontweight='bold')
plt.ylabel("Percentage")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_density = df.nlargest(10, 'Population density')

plt.figure(figsize=(10,5))
sns.barplot(x='Country', y='Population density', data=top_density, palette='mako')
plt.title("Top 10 Countries by Population Density", fontsize=13, fontweight='bold')
plt.ylabel("People per kmÂ²")
plt.xlabel("Country")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_young = df.nlargest(10, 'Population Aged 0 to 14 (%)')

plt.figure(figsize=(10,5))
sns.barplot(x='Country', y='Population Aged 0 to 14 (%)', data=top_young, palette='coolwarm')
plt.title("Top 10 Countries with Youngest Populations (Age 0â€“14 %)", fontsize=13, fontweight='bold')
plt.ylabel("Percentage of Youth Population")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_old = df.nlargest(10, 'Population Aged 60 and Over (%)')

plt.figure(figsize=(10,5))
sns.barplot(x='Country', y='Population Aged 60 and Over (%)', data=top_old, palette='crest')
plt.title("Top 10 Countries with Oldest Populations (Age 60+ %)", fontsize=13, fontweight='bold')
plt.ylabel("Percentage of Elderly Population")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='Country', y='Sex ratio (males per 100 females)', data=top_pop, palette='flare')
plt.title("Sex Ratio by Country â€” Top 10 Populated Nations", fontsize=13, fontweight='bold')
plt.ylabel("Males per 100 Females")
plt.axhline(100, color='gray', linestyle='--', linewidth=1)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap â€” Population Variables", fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```
